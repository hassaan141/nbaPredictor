```python
import pandas as pd
```

```python
#First load the data from a csv dataset
nbaData = pd.read_csv("nba_games.csv", index_col=0)
```

```python
nbaData
```

```python
#Sort by data
nbaData = nbaData.sort_values('date')
```

```python
#Reset index
nbaData = nbaData.reset_index(drop=True)
```

```python
nbaData
```

```python
#deleting not important rows
del nbaData["mp.1"]
del nbaData["mp_opp.1"]
del nbaData["index_opp"]
nbaData
```

```python
#over here, we create a new column called target, which gets the next games outcome
def add_target(team):
    team["nextGame"] = team["won"].shift(-1)
    return team

#Target is the next games match outcome, so we shift the next games match result into its own column
nbaData = nbaData.groupby("team", group_keys=False).apply(add_target)
```

```python
nbaData[nbaData["team"]=="WAS"]
```

```python
nbaData["nextGame"][pd.isnull(nbaData["nextGame"])]=2
```

```python
nbaData
```

```python
#This value gives a 1 or 0 value to the nextGame column to sognigy a win or a loss
nbaData["nextGame"] = nbaData["nextGame"].astype(int, errors="ignore")
```

```python
#this gives how many wins and losses there are bc always 1 team has to win or loose
nbaData["won"].value_counts()
```

```python
nbaData["nextGame"].value_counts()
```

```python
# What this does is that it changes all the not null values to false
nulls = pd.isnull(nbaData)
```

```python
#This gives the count of all the null values, or all the true values in each row
nulls = nulls.sum()
```

```python
#This gives us the how many null values are in each column
nulls
```

```python
#This gives us all the columns which have null values: nulls[nulls>0]
# We are now going to delete these values
nulls = nulls[nulls>0]
```

```python
nulls
```

```python
#This creates a list of columns that are not null ~ is a negation so it only puts the values that have values
valid_columns = nbaData.columns[~nbaData.columns.isin(nulls.index)]
```

```python
#now we want to assign the columns that are only valid  do the dataframe using .copy
nbaData = nbaData[valid_columns].copy()
```

```python
nbaData
```

```python
##############################################################################################
#All the above steps were data cleaning
#############################################################################################
```

```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
#We dont want to learn all the data, we want to train some data, then predicts the rest
#We split it up, we train on one set, then we predict another set
#   CROSS 
#Because we time ordered, we train past data to predict future data
# Use data from 2015 to predict 2022
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
#Our machine learning model, uses ridge regression to do classification
from sklearn.linear_model import RidgeClassifier

rr = RidgeClassifier(alpha = 1)
split = TimeSeriesSplit(n_splits=3)

#This picks the best feature for us, pick about 20% so 30 features, direction is forward because
#This is time sequenced
sfs = SequentialFeatureSelector(rr, n_features_to_select = 30, direction = "forward", cv=split)
```

```python
#Now we need to scale the data, scale all the columns so they fall between 0 and 1
# remove the columns that cannot be scaled
removed_columns = ["season", "date", "won", "nextGame", "team", "team_opp"]
```

```python
selected_col = nbaData.columns[~nbaData.columns.isin(removed_columns)]
```

```python
#Now we are ready to scale, this will scale the data between 0 and  1
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
#Why we do it like this is because we only want to scale the numerical values columns, so this way is good
nbaData[selected_col] = scaler.fit_transform(nbaData[selected_col])

#Makes rige regression perform better
```

```python
nbaData
```

```python
sfs.fit(nbaData[selected_col], nbaData["nextGame"])
```
