```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
#We dont want to learn all the data, we want to train some data, then predicts the rest
#We split it up, we train on one set, then we predict another set
#   CROSS 
#Because we time ordered, we train past data to predict future data
# Use data from 2015 to predict 2022
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
#Our machine learning model, uses ridge regression to do classification
from sklearn.linear_model import RidgeClassifier

rr = RidgeClassifier(alpha = 1)
split = TimeSeriesSplit(n_splits=3)

#This picks the best feature for us, pick about 20% so 30 features, direction is forward because
#This is time sequenced
sfs = SequentialFeatureSelector(rr, n_features_to_select = 30, direction = "forward", cv=split)
```

```python
#Now we need to scale the data, scale all the columns so they fall between 0 and 1
# remove the columns that cannot be scaled
removed_columns = ["season", "date", "won", "nextGame", "team", "team_opp"]
```

```python
selected_col = nbaData.columns[~nbaData.columns.isin(removed_columns)]
```

```python
#Now we are ready to scale, this will scale the data between 0 and  1
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
#Why we do it like this is because we only want to scale the numerical values columns, so this way is good
nbaData[selected_col] = scaler.fit_transform(nbaData[selected_col])

#Makes rige regression perform better
```

```python
sfs.fit(nbaData[selected_col], nbaData["nextGame"])
```

```python
#Get the list of our predictor columns out of our sequential selector
#If the feature selected has picked that column to use in our model, this will be true, else it will be false
predictors = list(selected_col[sfs.get_support()])
```

```python
#These are the 30 columns that we will use to train our data to make predictons
# This makes sense because the team will be a team will usually have similar stats to the other team
predictors
```

```python
#Will help us make the predictons
# we have time series data, we have to use historic data to predict future data, 
# will split by season to predict the past season to predict fututre season
# start parameter is 2 bc we need atleast 2 seasons of data to start making predictions
# will use 2016, 2017 data to make predictions for 2018
#Then we will use 16, 17, 18 to predict 2019
def backtest(data, model, predictors, start=3, step=1):
    all_predictions=[]

    seasons = sorted(data["season"].unique())

    for i in range (start, len(seasons), step):
        season = seasons[i]

        train = data[data["season"]< season]
        test = data[data["season"] == season]

        model.fit(train[predictors], train["nextGame"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)

        #Axis 1 will create the result as 2 seperate columns
        combined = pd.concat([test["nextGame"], preds], axis=1)
        combined.columns=["actual", "prediction"]

        all_predictions.append(combined)
    return pd.concat(all_predictions)

          
```

```python
# sorted(nbaData["season"].unique())
predictions= backtest(nbaData, rr, predictors)
```

```python
def find_team_averages(team):
    # Select only numeric columns for the rolling mean
    rolling = team.select_dtypes(include=[float, int]).rolling(10).mean()
    
    # Return the rolling means first, then append the non-numeric columns like 'team' and 'season'
    return pd.concat([rolling.reset_index(drop=True), team[['team', 'season']].reset_index(drop=True)], axis=1)

# Group by team and season, and apply the rolling function
nbaData_rolling = nbaData.groupby(["team", "season"]).apply(find_team_averages)

# Reset index if necessary after groupby
nbaData_rolling = nbaData_rolling.reset_index(drop=True)
```
